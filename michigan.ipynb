```python
# gensim modules
from gensim import utils
from gensim.models.doc2vec import LabeledSentence
from gensim.models import Doc2Vec
from collections import namedtuple
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.utils import to_categorical
from keras import models
from keras import layers




from gensim.models import Word2Vec , Doc2Vec
f = open ( '/home/farzad/Documents/kaggle/traindata.txt' , 'r');

labels =[];
sents=[];
for line in f.readlines():
    try:
        splited=line.split("\t")
        labels.append(int(splited[0]))
        sents.append(splited[1])
    except :
        pass
    
docs = []
analyzedDocument = namedtuple('AnalyzedDocument', 'words tags')
for i, text in enumerate(sents):
    words = text.lower().split()
    tags = [i]
    docs.append(analyzedDocument(words, tags))
```

```python
#modelDoc2Vec = Doc2Vec(min_count=1, window=10, vector_size=200, sample=1e-4, negative=5, workers=7)
modelDoc2Vec = Doc2Vec(vector_size=300, window=8, negative=5, hs=0, min_count=1, workers=8)

modelDoc2Vec.build_vocab(docs)
```

```python
len(modelDoc2Vec.docvecs[0])
x_train = []
for dc in range(len(modelDoc2Vec.docvecs)) :
    x_train.append(modelDoc2Vec[dc])
    
```

```python
# x_train = [item for sublist in x_train for item in sublist]
# len(x_train)
# type(x_train)
#x_train[0].shape

x_train = np.array(x_train)
type(x_train)
```

```python


model = Sequential() # Creating Model By Adding Layers One By One

# Input - Layer
model.add(layers.Dense(100, activation = "relu", input_shape=(300, )))

# Hidden - Layers
model.add(layers.Dropout(0.3, noise_shape=None, seed=None))
model.add(layers.Dense(100, activation = "relu"))
          
model.add(layers.Dropout(0.2, noise_shape=None, seed=None))
model.add(layers.Dense(100, activation = "relu"))

# Output- Layer
model.add(layers.Dense(1, activation = "sigmoid"))

model.summary()


model.compile(
 optimizer = "adam",
 loss = "binary_crossentropy",
 metrics = ["accuracy"]
)
results = model.fit(
 x_train, labels,
 epochs= 10,
 batch_size = 64,
 validation_split=.1
)
print("Test-Accuracy:", np.mean(results.history["val_acc"]))
```
